```python
from json import load
import pandas as pd
```

```python
with open("./TopTenTechniques.json", 'r') as json_file:
    top_ten_techniques = load(json_file)
```

```python
ttt_mitigations = list()
for technique in top_ten_techniques:
    ttt_mitigations.extend(technique['mitigations'])
ttt_mitigations
```

```python
unique_mid = set()
unique_ttt_mitigations = list()
for mitigation in ttt_mitigations:
    if mitigation['mid'] not in unique_mid:
        unique_mid.add(mitigation['mid'])
        mitigation['rating'] = len([*filter(lambda tmp: mitigation['mid'] == tmp['mid'], ttt_mitigations)])
        unique_ttt_mitigations.append(mitigation)
unique_ttt_mitigations
```

```python
df = pd.read_csv('Перечень сформированных угроз.csv')
df_fstec_mitigations = df['Меры защиты'].to_list()
```

```python
fstec_mitigations = list()
for mitigations in df_fstec_mitigations:
    fstec_mitigations.extend(map(lambda mitigation: dict(mid=mitigation[:mitigation.index(' ')],
                                                         description=mitigation[mitigation.index(' '):].strip(),
                                                         rating=0), mitigations.split(';\r\n')))
unique_mid = set()
unique_fstec_mitigations = list()
for mitigation in fstec_mitigations:
    if mitigation['mid'] not in unique_mid:
        unique_mid.add(mitigation['mid'])
        mitigation['rating'] = len([*filter(lambda tmp: mitigation['mid'] == tmp['mid'], fstec_mitigations)])
        unique_fstec_mitigations.append(mitigation)
unique_fstec_mitigations.sort(key=lambda tmp: tmp['rating'], reverse=True)
unique_fstec_mitigations
```
